```python
def status_types():
    barVulnerable = []
    barEndangered = []
    barCritEndangered = []
    barExtinct = []

    barWidth = 0.2
    for i in range(0, len(state_list)):
        barVulnerable.append(dstatus_state[('d' + (state_list[i]))]['Vulnerable'])
        barEndangered.append(dstatus_state[('d' + (state_list[i]))]['Endangered'])
        barCritEndangered.append(dstatus_state[('d' + (state_list[i]))]['Critically Endangered'])
        barExtinct.append(dstatus_state[('d' + (state_list[i]))]['Extinct'])

    r1 = np.arange(len(barVulnerable))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    r4 = [x + barWidth for x in r3]
    
    plt.bar(r1, barVulnerable, color='#ade6b1', width= barWidth, label='Vunerable')
    plt.bar(r2, barEndangered, color='#ade2e6', width= barWidth, label='Endangered')
    plt.bar(r3, barCritEndangered, color='#adc6e6', width= barWidth, label='Critically Endangered')
    plt.bar(r4, barExtinct, color='#e6ade2', width= barWidth, label='Extinct')
    
    
    plt.title("The Threatened Species Status of the States and Terriroties of Australia\n", fontsize=26,  family = 'monospace',)
    plt.xlabel("States and Territories of Australia", fontweight='bold', fontsize=22)
    plt.xticks([r + barWidth for r in range(len(barVulnerable))],['ACT', 'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC',
       'WA', 'ACI', 'CKI', 'CI', 'CSI', 'JBT', 'NFI', 'HMI', 'AAT', 'CMA'], fontsize=22)
    plt.ylabel("Number of Species", rotation='horizontal', fontweight='bold',fontsize=22, ha='right')
    plt.yticks(fontsize=16)
    
    plt.legend(fontsize=22)
    return plt.show
```

## Overall Endangered Status Score

By assigning a weighting to each key threatened status we can calculate an approximate overall endangered status score for each state.

### WEIGHTING OF EACH STATUS
**Vulnerable = 1, Endangered = 2, Critically Endangered = 3, Extinct = 4**

```python
def score_state(state2):
    state2_list = ((dstatus_state[('d' + (state2))]['Vulnerable']*1) + (dstatus_state[('d' + (state2))]['Endangered']*2) + (dstatus_state[('d' + (state2))]['Critically Endangered']*3) + (dstatus_state[('d' + (state2))]['Extinct']*4))
    return state2_list 
```

```python
list_scores = []
for i in range(0, len(state_list)):
    list_scores.append(score_state(state_list[i]))
print(list_scores)

def weighted_scores():
    y_pos = np.arange(len(state_list))
    
    plt.bar(y_pos,list_scores,color=colour_states1)
    
    plt.xticks(y_pos,state_list, fontsize=16)
    plt.xlabel('States and Territories of Australia', fontsize=22)
    plt.yticks(fontsize=16)
    plt.ylabel('Weighted Endangered Status Score', fontsize=22)
    return plt.show    
```
